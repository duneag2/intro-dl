```python
def make_linear(w=0.5, b=0.8, size=50, noise=1.0):
    x = np.random.rand(size)
    y = w * x + b
    noise = np.random.uniform(-abs(noise), abs(noise), size=y.shape)
    yy = y + noise
    plt.figure(figsize=(10, 7))
    plt.plot(x, y, color='r', label=f'y = {w}*x + {b}')
    plt.scatter(x, yy, label='data')
    plt.legend(fontsize=20)
    plt.show()
    print(f'w: {w}, b:{b}')
    return x, yy
```

```python
x, y = make_linear(w=0.3, b=0.5, size=100, noise=0.01)

y[5] = 0.75
y[10] = 0.75

plt.figure(figsize=(10, 7))
plt.scatter(x, y)
plt.show()
```

```python
w = np.random.uniform(low=-1.0, high=1.0)
b = np.random.uniform(low=-1.0, high=1.0)
```

```python
y_hat = w * x + b

error = (y_hat - y) ** 2
```

```python
##(1) Mean Squared Error(MSE) Loss
##(2) 'error = (y_hat - y)'    MSE loss function   .
## MSE             . Loss function          .
##'error = np.abs(y_hat - y)'(MAE) Absolute Error   outlier            ,        .
```

```python
num_epoch = 5000

learning_rate = 0.5
```

```python
##(1)Epoch refers to the one entire passing of training data through the algorithm.
##(2)Learning rate is a tuning parameter in an optimization algorithm that determines the step size at each iteration while moving toward a minimum of a loss function.
```

Question1

```python
errors = [] #    
w = np.random.uniform(low=-1.0, high=1.0) # w -1.0 1.0    
b = np.random.uniform(low=-1.0, high=1.0) # b -1.0 1.0    

for epoch in range(num_epoch):
    y_hat = x * w + b #     

    error = ((y_hat - y)**2).mean() #   
    if error < 0.0005:
        break #    0.0005   

    #     
    w = w - learning_rate * ((y_hat - y) * x).mean()
    b = b - learning_rate * (y_hat - y).mean()

    errors.append(error) #   

    if epoch % 5 == 0:
        print("{0:2} w = {1:.5f}, b = {2:.5f} error = {3:.5f}". format(epoch, w, b, error)) #epoch 5  ,  epoch, , ,  

#  
print("----" * 15)
print("{0:2} w = {1:.1f}, b = {2:.1f} error = {3:.5f}".format(epoch, w, b, error))
```
